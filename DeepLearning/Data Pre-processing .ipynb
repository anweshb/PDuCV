```python
import os
import shutil
import matplotlib.pyplot as plt
import math
```

```python
base_directory = "E:\PDuCV\chest_xray"
```

```python
for item in os.listdir(base_directory):
    print(item)
```

```python
train_directory = base_directory + "\\train"
val_directory = base_directory + "\\val"
test_directory = base_directory + "\\test"
```

```python
total_train = len(os.listdir(train_directory+"\\NORMAL")) + len(os.listdir(train_directory+"\\PNEUMONIA"))
total_val = len(os.listdir(val_directory+"\\NORMAL")) + len(os.listdir(val_directory+"\\PNEUMONIA"))
total_test = len(os.listdir(test_directory+"\\NORMAL")) + len(os.listdir(test_directory+"\\PNEUMONIA"))

print("Images in training directory : ",total_train)
print("Images in validation directory : ",total_val)
print("Images in test directory : ",total_test)
print("Total number of images: ",total_train+ total_test+ total_val)
```

```python
labels = "Training","Validation","Testing"
sizes = [total_train, total_val, total_test]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

# Considering the total amount of images the ideal split should be 70-15-15

```python
print("Ideal number of images in training set: ",math.floor(0.7*(total_train+ total_test+ total_val)))
print("Ideal number of images in validation set: ",math.floor(0.15*(total_train+ total_test+ total_val)))
print("Ideal number of images in test set: ",math.floor(0.15*(total_train+ total_test+ total_val)))
```

```python
print("Total number of NORMAL images in training set: ",len(os.listdir(train_directory+"\\NORMAL")))
print("Total number of PNEUMONIA images in training set: ",len(os.listdir(train_directory+"\\PNEUMONIA")))
```

```python
count = 0
for images in os.listdir(train_directory+"\\PNEUMONIA"):
    if count == 800:
        break
    else:
        shutil.move(train_directory+"\\PNEUMONIA\\"+images,val_directory+"\\PNEUMONIA")
        count += 1
```

```python
count = 0
for images in os.listdir(train_directory+"\\PNEUMONIA"):
    if count == 200:
        break
    else:
        shutil.move(train_directory+"\\PNEUMONIA\\"+images,test_directory+"\\PNEUMONIA")
        count += 1
```

```python
new_total_train = len(os.listdir(train_directory+"\\NORMAL")) + len(os.listdir(train_directory+"\\PNEUMONIA"))
new_total_val = len(os.listdir(val_directory+"\\NORMAL")) + len(os.listdir(val_directory+"\\PNEUMONIA"))
new_total_test = len(os.listdir(test_directory+"\\NORMAL")) + len(os.listdir(test_directory+"\\PNEUMONIA"))
```

```python
labels = "Training","Validation","Testing"
sizes = [new_total_train, new_total_val, new_total_test]
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0)

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
